```python
import numpy as np
import cv2
import matplotlib.image as mating
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# OpenCV - BGR (blue, green, red)
img_cv = cv2.imread('./data/alunos_01.jpg')
# Matplotlib - RGB (red, green, blue)
img_mat = mating.imread('./data/alunos_01.jpg')
# PIL - RGB, and return an object
img_pil = Image.open('./data/alunos_01.jpg')
```

```python
# OpenCV - BGR
print(img_cv)
```

```python
# matplotlib
print(img_mat)
```

```python
img_pil
```

```python
# converting PIL objet to array
img_pil_arr = np.array(img_pil)
print(img_pil_arr)
```

## Display Image

```python
# Display image using OpenCV
cv2.imshow('OpenCV BGR', img_cv)
cv2.imshow('Matplotlib RGB', img_mat)
cv2.imshow('Pillow RGB', img_pil_arr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Display image using matplotlib
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')

plt.subplot(1,3,2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')

plt.subplot(1,3,3)
plt.imshow(img_pil_arr)
plt.axis('off')
plt.title('Pillow RGB')

plt.show()
```

## Color Space
- Color Conversion
- Gray Scale

```python
img_cv.shape
```

```python
# numpy slicing
# BGR
b = img_cv[:,:,0]
g = img_cv[:,:,1]
r = img_cv[:,:,2]
```

```python
cv2.imshow('blue channel', b)
cv2.imshow('green channel', g)
cv2.imshow('red channel', r)

cv2.imshow('BGR', img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b, g, r = cv2.split(img_cv)
```

```python
b
```

```python
cv2.imshow('blue channel', b)
cv2.imshow('green channel', g)
cv2.imshow('red channel', r)

cv2.imshow('BGR', img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
b.shape, g.shape, r.shape, img_cv.shape
```

```python
# converting color
# bgr to rgb
img_cv_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cv_rgb)
plt.axis('off')
plt.show()
```

## GrayScale

```python
img_gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('grayscale', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
